## **Introduction**

**Domain of a Protein**  <a name="Domain"></a>

Domains are regions in a protein's polypeptide chain that are self-stabilizing and those fold independently from the rest. They are responsible for a particular function or interaction, contributing to the overall role of a protein. Uniprot defines domains as as a specific combination of secondary structures organized into a characteristic three-dimensional structure or fold. Many proteins consist of several domains, and a domain may appear in a variety of different proteins. In general, domains vary in length from between about 50 amino acids up to 250 amino acids in length. Domains often form functional units and  may exist in a variety of biological contexts, where similar domains can be found in proteins with different functions.




**Structural Domain of a Protein**  <a name="sf"></a>

Structural domains are specific parts of a protein that form a distinct three-dimensional structure, which is stable and can exist independently of the rest of the protein chain This means that even if a structural domain is isolated from the rest of the protein, it will still maintain its shape. Structural domains are often composed of stable combinations of secondary structures like alpha-helices and beta-sheets.These domains have clear boundaries within the protein, separating them from other domains or regions.

Some examples are :
- Immunoglobulin Fold: Found in antibodies, consisting of a beta-sandwich of two beta-sheets.
- TIM Barrel: A common fold found in many enzymes, consisting of eight alpha-helices and eight parallel beta-strands that alternate along the peptide backbone.




**Functional Domain of a Protein** <a name="functional"></a>

Functional domains are specific regions of a protein that carry out a particular function.They are responsible for a specific activity or interaction. For example, one domain might bind a substrate, while another might catalyze a chemical reaction. Functional domains often contain conserved amino acid sequences or motifs that are crucial for their function.A single protein can have multiple functional domains, each with a different role. This allows the protein to participate in various cellular processes.

Some examples are :

- SH2 (Src Homology 2) Domain: Binds specifically to phosphorylated tyrosine residues, playing a crucial role in signal transduction pathways.
- Kinase Domain: Found in protein kinases, responsible for transferring phosphate groups from ATP to specific substrates, which is a key mechanism in regulating cellular activities.
- DNA-binding Domains: Such as zinc fingers, helix-turn-helix, and leucine zipper domains, which enable proteins to interact with specific DNA sequences, often regulating gene expression.

In this tutorial we'll focus on functional domain of a protein.
Functional domains are significant because they define the specific activities and interactions that a protein can perform. These domains allow proteins to carry out complex biological functions by providing distinct regions that specialize in particular tasks.

Functional domains in proteins are crucial for several reasons:

**Specificity and Function:** They allow proteins to perform specific activities, such as catalyzing reactions, binding to other molecules, or interacting with DNA. This specificity is essential for the regulation and coordination of numerous cellular processes​.

**Modularity and Evolution:** They are often modular, meaning they can be recombined in different ways within proteins to create new functions. This modularity facilitates evolutionary adaptation by allowing new proteins to be formed from existing functional units​​.

**Conserved Mechanisms:**  Many functional domains are conserved across species, highlighting their fundamental roles in biology. This conservation allows researchers to study these domains in model organisms and apply their findings to human biology​​.

**Protein Interactions:** Functional domains mediate interactions between proteins and other molecules, which is vital for processes like signal transduction, gene regulation, and metabolic control​.

**Enzymatic Activity:** Functional domains often possess catalytic activities necessary for metabolism, DNA replication, and other critical cellular functions​.

**Importance of Functional Domain of Protein**

Understanding of functional domains is crucrial for target identification and drug design.

- Functional domains help identify potential drug targets. For example, kinases and proteases are common drug targets because their functional domains are well-defined.
- Drugs can be designed to specifically bind and inhibit the activity of a functional domain, such as kinase inhibitors targeting the ATP-binding site of kinase domains​.


In this tutorial we'll predict functional domain of protein for drug target slection. Functional domain prediction plays a significant role in drug target validation by providing valuable insights into the protein's function and potential druggable sites. Here's how it contributes:



1. Understanding Protein Function: By identifying these functional domains OF protein, we can gain a deeper understanding of the protein's overall role in the cell. This knowledge helps assess if targeting the protein with a drug would be a valid therapeutic strategy for the disease of interest.

2. Prioritizing Druggable Sites:  Many functional domains within a protein are directly involved in its essential activities. Targeting these domains with drugs might disrupt the protein's normal function and cause unintended consequences. Functional domain prediction helps researchers differentiate between essential and non-essential domains. Drugs can then be designed to target the non-essential domains or the interfaces between domains, potentially minimizing disruption of the protein's core functionality.


3. Informing Combination Therapies:  Some diseases may require targeting multiple proteins or pathways for effective treatment. Functional domain prediction can help identify proteins with distinct but complementary functions that contribute to the disease. By targeting these different functional domains with a combination of drugs, researchers can potentially achieve a more synergistic and effective therapeutic effect.

**Understanding Protein Functional Domains through GO Terms** <a name="GO"></a>


Gene Ontology (GO) provides a structured and controlled vocabulary to describe gene and gene product attributes in any organism. It consists of three main sub-ontologies: Molecular Function (MF), Biological Process (BP), and Cellular Component (CC). Each sub-ontology provides a different perspective on the roles of genes and function of proteins.


**Molecular Function (MF)**

Molecular Function GO terms describe specific activities of protein domains, such as binding or catalytic activities. For example, a kinase domain might be annotated with GO:0004672 (protein kinase activity), providing a precise functional characterization of what the domain does at a molecular level​.For instance, the SH2 domain, which binds phosphorylated tyrosines, would be associated with a molecular function GO term related to binding activity.

 Some other examples:

- Transporter Activity: Enabling the movement of substances across cell membranes (GO:0005215).
- Catalytic Activity: Acting as an enzyme to speed up biochemical reactions (GO:0003824).
- Binding Activity: Interacting selectively and non-covalently with another molecule (GO:0005488).




**Biological Process (BP)**

Biological Process GO terms help identify the biological processes in which a protein domain is involved. This can reveal the broader biological role of the domain, such as signal transduction, cell cycle regulation, or metabolic processes. For example: A domain involved in signal transduction might be annotated with GO:0007165 (signal transduction), indicating its role in the cellular communication network.

 Some other examples:

- Cellular Process: Any process carried out at the cellular level (GO:0009987).
- Signal Transduction: The process by which a signal is conveyed to trigger - cellular responses (GO:0007165).
- Transport: The directed movement of substances within or between cells (GO:0006810).



**Cellular Component (CC)**

GO terms specify where within the cell the protein domain functions. This is crucial for understanding the context in which the domain operates, such as within membranes, organelles, or specific cellular structures.For example: The transmembrane domain might be associated with GO:0016020 (membrane), indicating its role in membrane-spanning functions.


 Some other examples:
- Membrane: A lipid bilayer surrounding cells or organelles (GO:0016020).
- Cell Periphery: The outer part of the cell, including the plasma membrane (GO:0071944).
- Ribosome: The complex that synthesizes proteins in the cell (GO:0005840).




**How to use the predicted Gene Ontology for analysis of target protein**

**1. Functional Analysis**
* **Biological Relevance**:
Examine the GO terms to understand the biological processes and pathways the potential target protein is involved in. This helps in determining the relevance of the protein to the disease mechanism.

* **Molecular Functions:** Assess the molecular functions of the protein, such as enzyme activity, binding properties, or receptor activity, which can provide insights into how the protein might be modulated by a drug.
* **Cellular Localization:** Identify where in the cell the protein operates, which is crucial for understanding its role in cellular pathways and potential interactions with other proteins.

**2. Pathway Analysis**

* **Pathway Mapping:** Map the GO terms to known biological pathways to see how the potential drug target fits into these pathways. Tools like KEGG or Reactome can be used for this purpose.
* **Interaction Networks:** Build protein-protein interaction networks using GO annotations to visualize how the potential target interacts with other proteins and pathways.

**3. Predicting Off-Target Effects**

* **Cross-Referencing GO Terms:** Use GO terms to predict off-target effects by identifying other proteins that share similar functions or participate in the same pathways. This helps in assessing potential side effects of the drug.
* **Functional Similarity:** Analyze the functional similarity between the target protein and other proteins to identify any unintended interactions or effects.

**4. Experimental Validation**

* **Hypothesis Generation:** Generate hypotheses about the role of the target protein in the disease based on its GO annotations. This guides the design of experiments to validate the target.

## **Predicting Functional Domain of a Protein** <a name="pfp"></a>

**Domain-PFP**

Proteins consist of domains, which are functional and structural units responsible for specific functions and interactions. Therefore, it is compelling to infer the functions of a protein based on the presence and distribution of the various domains in it. Domain-PFP is a self-supervised method to learn functional representations of protein domains that can be used for protein function prediction.


**Learning Domain Embeddings**

The overall pipeline for learning functionally aware domain embeddings in the Domain-PFP method begins with the collection of protein sequences from databases like Swiss-Prot, followed by the assignment of InterPro domains to these sequences using InterProScan. Concurrently, Gene Ontology (GO) terms are gathered from UniProt. The conditional probabilities of domain-GO co-occurrences are calculated from the dataset, which serve as pseudo labels for the self-supervised learning process. These probabilities represent how likely a protein containing a specific domain will have a particular GO function.

A neural network is then trained to learn 256-dimensional embeddings for domains and GO terms. This network computes the Hadamard product of the domain and GO embeddings and passes them through densely connected layers to predict the conditional probabilities of domain-GO associations. The training process includes regularization techniques like dropout and L1 regularization to enhance model performance and sparsity. Negative sampling is employed to address class imbalance by including a large number of non-co-occurring domain-GO pairs. The network is trained and validated on separate data splits, and the best model is selected based on validation performance.

Once the domain embeddings are learned, they are used to represent proteins by averaging the embeddings of all domains present in each protein. These protein embeddings are then used in a K-Nearest Neighbors (KNN) classifier to predict protein functions. The overall pipeline is validated by comparing the model's performance against existing methods using benchmarks like PROBE and CAFA3, with metrics such as Fmax, AUPR, and Smin demonstrating its effectiveness. This approach ensures the creation of functionally aware domain embeddings that significantly improve the accuracy of protein function predictions, making it a valuable tool in areas such as drug target validation.

**Domain Embedding model**

```python
import numpy as np
import torch
torch.manual_seed(2)
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import os



class DomainGOEmbeddingModel(torch.nn.Module):
    """
    Domain GO Embedding Model
    """

    def __init__(self, domain_mapper, go_mapper, emb_dim=256, lmbd=0.1, n_neuron_1 = 128, dp1=0.05, name='mdl'):
        """
        Initializes the model

        Args:
            domain_mapper (dict): python dictionary mapping Domain to id.
            go_mapper (dict): python dictionary mapping GO term to id.
            emb_dim (int): dimension of Embedding.
            lmbd (float): lambda in L1 regularization. Defaults to 0.1.
            n_neuron_1 (int, optional): number of neurons in dense layer. Defaults to 128.
            dp1 (float, optional): dropout rate. Defaults to 0.05.
            name (str, optional): name of model. Defaults to 'mdl'.
        """
        super(DomainGOEmbeddingModel, self).__init__()

        self.name = name

        self.emb_dim = emb_dim
        self.lmbd = lmbd

        self.num_domains = len(domain_mapper)
        self.domain_mapper = domain_mapper
        self.embedding_domain = torch.nn.Embedding(num_embeddings=self.num_domains, embedding_dim=self.emb_dim)

        self.num_go = len(go_mapper)
        self.go_mapper = go_mapper
        self.embedding_go = torch.nn.Embedding(num_embeddings=self.num_go, embedding_dim=self.emb_dim)

        self.dense1 = torch.nn.Linear(emb_dim, n_neuron_1)
        self.act1 = torch.nn.ReLU()
        self.dp1 = torch.nn.Dropout(dp1)

        self.out = torch.nn.Linear(n_neuron_1,1)


    def forward(self, domain_id, go_id):
        """
        Implements forward pass

        Args:
            domain_id (int): id of the domain
            go_id (int): id of the GO term

        Returns:
            _type_: _description_
        """

        domain_embedding = self.embedding_domain(domain_id)
        go_embedding = self.embedding_go(go_id)

        feat = torch.mul(domain_embedding, go_embedding)

        feat = self.dense1(feat)
        feat = self.act1(feat)
        feat = self.dp1(feat)

        out = self.out(feat)

        return out, torch.mean(torch.abs(domain_embedding))     # return the score and absolute value of embeddings layer for L1 regularization


    def get_domain_embedding(self, domain_id):
        """
        returns the embedding of a domain

        Args:
            domain_id (int): id of the domain

        Returns:
            tensor: domain embedding
        """

        dmn_embdng = self.embedding_domain(domain_id)

        return dmn_embdng

    def get_go_embedding(self, go_id):
        """
        returns the embedding of a GO terms

        Args:
            go_id (int): id of the GO term

        Returns:
            tensor: GO term embedding
        """
        go_embdng = self.embedding_go(go_id)

        return go_embdng



def load_domaingo_embedding_model_weights(mdl, mdl_weight_pth):
    """
    loads the weights of a domaingo embedding model

    Args:
        mdl (torch model): model
        mdl_weight_pth (string): path to the saved weights

    Returns:
        torch model: model with the loaded weights
    """


    # since the model is computationally less expensive, we load it in cpu
    mdl.load_state_dict(torch.load(mdl_weight_pth,map_location=torch.device('cpu')))
    mdl.eval()

    return mdl
```

**Domain Embedding**

```python
import torch
torch.manual_seed(0)
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

DOUBLE_PRECISION = 12


class DomainEmbedding:
    """
    Domain Embedding Object
    """
    def __init__(self, domaingo_mdl, domain_mapper):
        """
        Initialization

        Args:
            domaingo_mdl (torch model): Domain GO Embedding Model
            domain_mapper (dict): maps domain to id

        """
        self.domain_mapper = domain_mapper
        self.total_domain = len(domain_mapper)

                            # extract the domain embeddings from the model
        self.embedding = np.array(domaingo_mdl.get_domain_embedding(torch.tensor(np.arange(self.total_domain))).detach().numpy(),dtype=np.float64)

        self.embedding = np.round(self.embedding,DOUBLE_PRECISION)


                                        # Normalize domain embeddings
        scaler = MinMaxScaler()
        self.embedding = scaler.fit_transform(self.embedding)
        self.embedding = np.round(self.embedding,DOUBLE_PRECISION)


    def get_embedding(self, inp):
        """
        returns the embedding of the input domain

        Args:
            inp (int): id of the input domain
                or
            inp (str): the input domain

        Returns:
            numpy array: embedding of the domain
        """

        if isinstance(inp,str):
            if inp in self.domain_mapper:
                return np.array(self.embedding[self.domain_mapper[inp]])
            else:                         # if domain not found, return 0 default
                return np.array(self.embedding[0]*0)

        else:
            if inp<0:                               # dummy value -1 can be used to obtain default 0 embedding
                return np.array(self.embedding[0]*0)

            if inp<len(self.domain_mapper):
                return np.array(self.embedding[inp])

            else:                         # if domain not found, return 0 default
                return np.array(self.embedding[0]*0)


    def contains(self, inp):
        """
        check if a particular domain is in the model

        Args:
            inp (int): id of the input domain
                or
            inp (str): the input domain

        Returns:
            bool: domain exists or not
        """
        if isinstance(inp,str):
            return inp in self.domain_mapper
        else:
            return inp < len(self.embedding)
```

**Download protein sequence**

```python
import requests
import os
from tqdm import tqdm
import sys

def download_sequence(pid):
    """
    Download protein sequence as fasta file

    Args:
        pid (str): UniProt ID of protein
    """

    try:
        os.makedirs('temp_data')
    except:
        pass

    r = requests.get(f'https://rest.uniprot.org/uniprotkb/{pid}.fasta')

    if(r.status_code==200):
        raw_fasta = r.text

        if(len(raw_fasta)==0):
            print('This UniProt ID is either invalid or obsolete', file=sys.stderr)
            return False

        fp = open(os.path.join('temp_data',pid+'.fasta'),'w')
        fp.write(raw_fasta)
        fp.close()
        return True

    else:
        print(r.text, file=sys.stderr)
        return False
```

```python
from argparse import ArgumentParser
from datetime import datetime
import pickle
import sys, os
sys.path.append(os.path.abspath('DomainPFP'))
import numpy as np
import pandas as pd
pd.set_option('display.max_colwidth', None)

DOUBLE_PRECISION = 12       # precision for floating point
```

```python
def parse_domains(tsv_file_pth):
    """
    Parses the domains from the InterPro Scan results

    Args:
        tsv_file_pth (str): path to the tsv file
    """

    domains = []

    with open(tsv_file_pth,'r') as fp:
        dt = fp.read().split('\n')[:-1]
        for ln in dt:
            domains.append(ln.split('\t')[11])

    domains = set(domains)-set({'-'})

    return domains
```

```python
def compute_embeddings(domains):
    """
    Computes the protein embedding from the domains

    Args:
        domains (set or list): set or list of domains

    Returns:
        tuple of 3 numpy arrays: (MF embedding, BP embedding, CC embedding)
    """

    domain_mapper_mf = pickle.load(open(os.path.join('data','processed','domain_mapper_swissprot_mf.p'),'rb'))      # loading the mapper files
    go_mapper_mf = pickle.load(open(os.path.join('data','processed','go_mapper_swissprot_mf.p'),'rb'))
    mdl_mf = DomainGOEmbeddingModel(domain_mapper_mf,go_mapper_mf)                                              # creating a model
    mdl_mf = load_domaingo_embedding_model_weights(mdl_mf, os.path.join('saved_models','swissprot_mf'))             # loading the model weights
    dmn_embedding_mf = DomainEmbedding(mdl_mf, domain_mapper_mf)                                                # creating the Domaing Embedding object

    domain_mapper_bp = pickle.load(open(os.path.join('data','processed','domain_mapper_swissprot_bp.p'),'rb'))
    go_mapper_bp = pickle.load(open(os.path.join('data','processed','go_mapper_swissprot_bp.p'),'rb'))
    mdl_bp = DomainGOEmbeddingModel(domain_mapper_bp,go_mapper_bp)
    mdl_bp = load_domaingo_embedding_model_weights(mdl_bp, os.path.join('saved_models','swissprot_bp'))
    dmn_embedding_bp = DomainEmbedding(mdl_bp, domain_mapper_bp)

    domain_mapper_cc = pickle.load(open(os.path.join('data','processed','domain_mapper_swissprot_cc.p'),'rb'))
    go_mapper_cc = pickle.load(open(os.path.join('data','processed','go_mapper_swissprot_cc.p'),'rb'))
    mdl_cc = DomainGOEmbeddingModel(domain_mapper_cc,go_mapper_cc)
    mdl_cc = load_domaingo_embedding_model_weights(mdl_cc, os.path.join('saved_models','swissprot_cc'))
    dmn_embedding_cc = DomainEmbedding(mdl_cc, domain_mapper_cc)


    mf_embedding = dmn_embedding_mf.get_embedding(-1)           # Initialize embeddings
    bp_embedding = dmn_embedding_bp.get_embedding(-1)
    cc_embedding = dmn_embedding_cc.get_embedding(-1)

    cnt = 0
    for dmn in domains:
        if dmn_embedding_mf.contains(dmn):
            mf_embedding += dmn_embedding_mf.get_embedding(dmn)
            cnt += 1
    if(cnt>1):
        mf_embedding /= cnt                     # averaging


    cnt = 0
    for dmn in domains:
        if dmn_embedding_bp.contains(dmn):
            bp_embedding += dmn_embedding_bp.get_embedding(dmn)
            cnt += 1
    if(cnt>1):
        bp_embedding /= cnt

    cnt = 0
    for dmn in domains:
        if dmn_embedding_cc.contains(dmn):
            cc_embedding += dmn_embedding_cc.get_embedding(dmn)
            cnt += 1
    if(cnt>1):
        cc_embedding /= cnt


    mf_embedding = np.round(mf_embedding,DOUBLE_PRECISION)
    bp_embedding = np.round(bp_embedding,DOUBLE_PRECISION)
    cc_embedding = np.round(cc_embedding,DOUBLE_PRECISION)

    return mf_embedding, bp_embedding, cc_embedding
```

```python
def main():
    """
    Computes the embeddings of a query protein
    """

    protein = ''
    fasta = ''
    savefile = ''

    if args.protein:
        protein = args.protein

    if args.fasta:
        fasta = args.fasta

    if args.savefile:
        savefile = args.savefile


    if len(protein)>0:
        print(f'Downloading sequence of {protein} from UniProt')
        flg = download_sequence(protein)
        if(not flg):
            sys.exit()
        else:
            fasta = os.path.join('temp_data',protein+'.fasta')


    elif len(fasta)>0:
        print(f'Loading the fasta file {fasta}')
        if(not os.path.isfile(fasta)):
            sys.exit(f'{fasta} file not found')

    else:
        sys.exit('Please input a protein UniProt ID or path to a fasta file')

    if len(savefile)==0:
        sys.exit('Please input a path to save the protein embeddings')

    try:
        os.makedirs('temp_data',exist_ok=True)
    except Exception as e:
        print(e)


    job_id = datetime.now().strftime("%H%M%S%f")

    print('Extracting protein domains using InterPro Scan')
    os.system(f'python3 DomainPFP/iprscan5.py --email domainpfp@gmail.com --sequence {fasta} --outfile temp_data/{job_id}')
    print("Domains Computed")
    domains = parse_domains(os.path.join('temp_data',f'{job_id}.tsv.tsv'))

    print("Computing Embeddings")
    mf_embedding, bp_embedding, cc_embedding = compute_embeddings(domains)

    print(f"Saving Embeddings as {savefile}")
    pickle.dump({'mf':mf_embedding,
                 'bp':bp_embedding,
                 'cc':cc_embedding},
                open(savefile,'wb'))
```
