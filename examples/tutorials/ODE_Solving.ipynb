```python
from deepchem.utils.differentiation_utils.integrate.ode import solver_euler_1_2
import numpy as np

def lotka_volterra(x, y):
    y1, y2 = y[0][0], y[1][0]

    return [np.array(1.1 * y1 - 0.4 * y1 * y2), np.array(0.1 * y2 * y1 - 0.4 * y2)]

sol = solver_euler_1_2(lotka_volterra, 0.0, [[10], [10]], 0.1, 50)
```

```python
import matplotlib.pyplot as plt

plt.plot(sol[0], sol[1][0])
plt.plot(sol[0], sol[1][1])
```
